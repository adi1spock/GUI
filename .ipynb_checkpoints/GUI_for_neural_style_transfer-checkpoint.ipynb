```python
import tkinter as tk
from PIL import ImageTk,Image
from imutils.video import VideoStream
from imutils import paths
import itertools
import imutils
import time
import cv2
```

```python
model1="models/instance_norm/candy.t7"
model2="models/instance_norm/composition_vii.t7"
model3="models/instance_norm/feathers.t7"
model4="models/instance_norm/la_muse.t7"
model5="models/instance_norm/mosaic.t7"
model6="models/instance_norm/starry_night.t7"
model7="models/instance_norm/the_scream.t7"
model8="models/instance_norm/udnie.t7"
model9="models/eccv16/the_wave.t7"
model_list=[model1,model2,model3,model4,model5,model6,model7,model8,model9]

def hello(x,model):
    if(x==1):
        
        modelPath=model;
        print("[INFO] loading style transfer model...")
        net = cv2.dnn.readNetFromTorch(modelPath)
        print("[INFO] starting video stream...")
        vs = VideoStream(src=0).start()
        time.sleep(2.0)
        while True:
            frame = vs.read()
            frame = imutils.resize(frame, width=600)
            orig = frame.copy()
            (h, w) = frame.shape[:2]
            blob = cv2.dnn.blobFromImage(frame, 1.0, (w, h),
                (103.939, 116.779, 123.680), swapRB=False, crop=False)
            net.setInput(blob)
            output = net.forward()
            output = output.reshape((3, output.shape[2], output.shape[3]))
            output[0] += 103.939
            output[1] += 116.779
            output[2] += 123.680
            output /= 255.0
            output = output.transpose(1, 2, 0)
            cv2.imshow("Input", frame)
            cv2.imshow("Output", output)
            key = cv2.waitKey(1) & 0xFF
            if key == ord("n"):
                (modelID, modelPath) = next(modelIter)
                net = cv2.dnn.readNetFromTorch(modelPath)
            elif key == ord("q"):
                break

        cv2.destroyAllWindows()
        cv2.waitKey(1)
        cv2.waitKey(1)
        cv2.waitKey(1)
        cv2.waitKey(1)
        vs.stop()
```

GUI CODE BELOW USING TKINTER MODULE

```python
root=tk.Tk()
t=1
def forward(x):
    global my_label
    global button_forward
    global button_backward
    global button_start
    global t
    t=t+1
    
    mylabel.grid_forget()
    my_label=tk.Label(image=img_list[x-1])
    button_forward=tk.Button(root,text=">>",command=lambda: forward(x+1))
    button_backward=tk.Button(root,text="<<",command=lambda: back(x-1))
    if(x==9):
        button_forward=tk.Button(root,text=">>",state=tk.DISABLED)
        
    my_label.grid(row=0,column=0,columnspan=4)
    button_forward.grid(row=1,column=1)
    button_backward.grid(row=1,column=0)

    
def back(x):
    global my_label
    global button_forward
    global button_backward
    global button_start
    global t
    t=t-1
    mylabel.grid_forget()
    my_label=tk.Label(image=img_list[x-1])
    button_forward=tk.Button(root,text=">>",command=lambda: forward(x+1))
    button_backward=tk.Button(root,text="<<",command=lambda: back(x-1))
    #button_start=tk.Button(root,text="START",command=lambda: start(x-1))
    if(x==1):
        button_backward=tk.Button(root,text="<<",state=tk.DISABLED)
        
    my_label.grid(row=0,column=0,columnspan=4)
    button_forward.grid(row=1,column=1)
    button_backward.grid(row=1,column=0)
    
def start():
    
    y=model_list[t-1]
    hello(1,y)
    
my_img1=ImageTk.PhotoImage(Image.open("images/candy.jpg"))
my_img2=ImageTk.PhotoImage(Image.open("images/composition_vii.jpg"))
my_img3=ImageTk.PhotoImage(Image.open("images/feathers.jpg"))
my_img4=ImageTk.PhotoImage(Image.open("images/la_muse.jpg"))
my_img5=ImageTk.PhotoImage(Image.open("images/mosaic.jpg"))
my_img6=ImageTk.PhotoImage(Image.open("images/starry_night_crop.jpg"))
my_img7=ImageTk.PhotoImage(Image.open("images/the_scream.jpg"))
my_img8=ImageTk.PhotoImage(Image.open("images/udnie.jpg"))
my_img9=ImageTk.PhotoImage(Image.open("images/wave_crop.jpg"))

img_list=[my_img1,my_img2,my_img3,my_img4,my_img5,my_img6,my_img7,my_img8,my_img9]
mylabel=tk.Label(image=img_list[0])
mylabel.grid(row=0,column=0,columnspan=4)

button_forward=tk.Button(root,text=">>",command=lambda: forward(2))
button_backward=tk.Button(root,text="<<",command=lambda: back(1))
button_start=tk.Button(root,text="START",command=lambda: start())
button_exit=tk.Button(root,text="EXIT",command=root.quit)


button_forward.grid(row=1,column=1)
button_backward.grid(row=1,column=0)
button_start.grid(row=1,column=2)
button_exit.grid(row=1,column=3)

#root.geometry("640x520")
root.mainloop()

```
